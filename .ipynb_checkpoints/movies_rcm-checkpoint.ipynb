## Creating handcrafted features

```python
movies = movies_df[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.info()
```

```python
movies.iloc[0].genres
```

```python
movies.iloc[0].keywords
```

```python
movies.loc[0, 'crew']
```

```python
def convert(obj):
    array = []
    for i in ast.literal_eval(obj):
        array.append(i['name'])
    return array
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    array = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            array.append(i['name'])
            counter += 1
        else:
            break
    return array
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    array = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            array.append(i['name'])
    return array
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ', '') for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.loc[0, 'tags']
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df.head()
```

## Preprocessing tags

lowercase, remove stopwords, vectorazition, tokenization, stemming

```python
new_df['tags'] = new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df.head()
```

```python
new_df.iloc[0].tags
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
len(cv.get_feature_names())
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    stemmed=[]
    for i in text.split():
        stemmed.append(ps.stem(i))
    return ' '.join(stemmed)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

# Calculate the Cosine similarity

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
#     print(movie_index)
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1])[1:6]
    for i in movies_list:
#         print(i)
        print(new_df.iloc[i[0]].title)
```

# Some results

```python
recommend('Independence Day')
```

```python
recommend('Alien')
```

```python
recommend('Iron Man')
```

```python
recommend('The Avengers')
```
